## Outliers Sales

```python
df.Sales.describe()
```

### STD

```python
sales_limit_up = df.Sales.mean() + (3 * df.Sales.std())
sales_limit_up
```

```python
sales_limit_lp = df.Sales.mean() - (3 * df.Sales.std())
sales_limit_lp
```

```python
outliers = df[(df.Sales > sales_limit_up) | (df.Sales < sales_limit_lp)]
outliers
```

### Z-score

```python
outliers_Z = df[np.abs((df.Sales - df.Sales.mean()) / df.Sales.std()) > 3]
outliers_Z
```

### delete outliers

```python
df.shape
```

```python
df.drop(outliers.index , inplace= True)
```
